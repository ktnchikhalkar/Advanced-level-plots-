```python
df = pd.DataFrame(index= ['John','Michael','Ross','Joey','Dwight','Jimmy','Pam'],
                  data={'Apples':[10,12,15,9,4,11,8], 'Oranges':[2,4,5,7,1,2,4]})
```

```python
df
```

```python
df.plot.bar(figsize=(15,4))
```

```python
df.plot.scatter('Apples', 'Oranges')
```

```python
ax= df.plot.kde(figsize=(15,3))
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(1, 3, figsize=(15,3))
fig.suptitle('Two variable plots')
df.plot.line(ax=ax1)
df.plot.scatter('Apples','Oranges',ax=ax2)
df.plot.bar(ax=ax3)
```
